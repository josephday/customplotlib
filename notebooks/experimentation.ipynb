```python
import pandas as pd
import numpy as np
from matplotlib import font_manager as fm
```

```python
from customplotlib.base import Customplotlib
plt = Customplotlib()
```

```python
plt.plot(x=[1.0,2.0,3.0], y=[1.0,1.5,3.0], label='That')#, color='strong-red')
plt.plot(x=[1.0,2.0,3.0], y=[3.0,2.0,1.0], label='This')#, color='red')
```

```python
test_data = pd.DataFrame({'ind': [1,2,3,1,2,3],
                          'dep': [3,2,1,1,1.5,3],
                          'label':['This', 'This', 'This', 'That', 'That', 'That']})
```

```python
plt.plot(x='ind', y='dep', data=test_data, label='label')
```

```python
ind = test_data['ind']
dep = test_data['dep']
label = test_data['label']
plt.plot(x=ind, y=dep, label=label)
```
